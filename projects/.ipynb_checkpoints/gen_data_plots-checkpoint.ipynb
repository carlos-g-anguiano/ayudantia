```python
# Following https://stackoverflow.com/a/64604031
import sys, os

from pathlib import Path

# in jupyter (lab / notebook), based on notebook path
module_path = str(Path.cwd().parents[0] / "py")

if module_path not in sys.path:
    sys.path.append(module_path)
```

```python
import matplotlib.pyplot as plt
import plot_tools
import zfit
import customPDFs
import numpy as np
import pandas as pd
```

```python
import SLSQP_zfit
```

```python
data_path = str(Path.cwd().parents[0] / "GenData" )
```

```python
cos = zfit.Space('cosThetaKMu', [-1,1])
```

```python
cos_np = np.linspace(-1, 1, 1000)
```

```python
AFB = zfit.Parameter('AFB', 0)
FH = zfit.Parameter('FH', 0.002)
```

```python
decay_rate = customPDFs.decayWidth(AFB, FH, cos)
```

```python
import SLSQP_zfit
```

```python
constraints = SLSQP_zfit.create_constraint(decay_rate)
SLSQP = SLSQP_zfit.SLSQP()
```

# Ajustes

## Bin -1

```python
data_bin_n1 = pd.read_csv(data_path + "/Bin-1.csv")
```

```python
plt.hist(data_bin_n1.cosThetaKMu, bins=20, range=[-1,1])
```

```python
# Pasando los datos a zfit.Data

zdata_bin_n1 = zfit.Data.from_pandas(pd.DataFrame(data_bin_n1.cosThetaKMu), cos)
```

```python
nll = zfit.loss.UnbinnedNLL(decay_rate, zdata_bin_n1)
```

```python
nll.value()
```

```python
AFB.set_value(0), FH.set_value(0.002)

minimum = SLSQP.minimize(nll)
mz = minimum.hesse()
```

```python
mz
```

```python
# Adding SLSQP to dictionary
# Data to be written
dictionary = {
    "-1" : {
        "NLL": minimum.fmin,
        "SLSQP" : {
            "AFB": AFB.value().numpy(),
            "errAFB": mz[AFB]['error'],
            "fh": FH.value().numpy(),
            "errFH": mz[FH]['error'],
        }
    }
}
```

```python
AFB.set_value(0), FH.set_value(0.002)

MIN = zfit.minimize.Minuit()
minimum = MIN.minimize(nll)
mz = minimum.hesse()
```

```python
mz
```

```python
dictionary['-1']['Minuit'] = {
    "AFB": AFB.value().numpy(),
    "errAFB": mz[AFB]['error'],
    "fh": FH.value().numpy(),
    "errFH": mz[FH]['error'],
}
```

```python
dictionary
```

```python
fig = plt.figure()
axes = plot_tools.create_axes_for_pulls(fig)
plot_tools.plot_model(data_bin_n1.cosThetaKMu, decay_rate, bins=20, 
                      axis=axes[0], pulls=True,
                      axis_pulls=axes[1], chi_x=0.05, chi_y=0.9)
axes[0].text(0.3, 0.1, 'AFB = 0', transform=axes[0].transAxes, fontsize=20)
axes[0].text(0.6, 0.1, 'FH = 0.002', transform=axes[0].transAxes, fontsize=20)
plt.show()
```

```python
del(zdata_bin_n1)
del(data_bin_n1)
```
